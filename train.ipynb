# Training the model

### Preparing the dataset

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# our dataset is in parquet format
file_path = "voraus-ad-dataset-100hz.parquet"
df = pd.read_parquet(file_path)

print(f"Dataset Loaded!")
```

### Viewing dataset

```python
# Look at the first 5 rows and all column names
print(df.head())
print("-" * 30)
print("All Column Names:")
print(df.columns.tolist())
```

### Filtering wanted sensors

Too many features (columns)!

Selecting our wanted features: sensor columns...

```python
# the sensors we want
sensors = [
    'robot_current', 'robot_voltage', 'system_current',
    'motor_torque_1', 'motor_torque_2', 'motor_torque_3', 
    'motor_torque_4', 'motor_torque_5', 'motor_torque_6',
    'joint_velocity_1', 'joint_velocity_2', 'joint_velocity_3', 
    'joint_velocity_4', 'joint_velocity_5', 'joint_velocity_6'
]

# filtering dataframe to only these sensors
sensor_data = df[sensors].values

# normalising
scaler = StandardScaler()
scaled_data = scaler.fit_transform(sensor_data)

# create the 300 number vectors
window_size = 20
samples = []

for i in range(0, len(scaled_data) - window_size):
    window = scaled_data[i : i + window_size]
    samples.append(window.flatten())

X_train = np.array(samples)
print(f"Final Input Shape: {X_train.shape}") # should be (Samples, 300)

print("Sensors selected for the 300-number vector:")
for i, name in enumerate(sensors):
    print(f"{i+1}: {name}")
```
